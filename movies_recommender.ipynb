which columns to take for our recommendation system? we are trying to create system using tags so as popularity is a numeric measure
it won't be a good fit , inshort numeric columns won't work for this type of project where we use tags
=> genres , id , keywords , title  , overview , cast , crew

```python
movies = movies[['movie_id' , 'title' , 'genres' , 'overview' , 'keywords' , 'cast' , 'crew']]
```

```python
movies.head()
```

we need three columns i.e. movie_id , title , tags
-> merge genres , overview , keywords , cast , crew to the tags column

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# first we need to convert string of list to list
import ast
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0].cast
```

```python
def convert_cast(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['crew'][0]
```

```python
def get_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
# overview is a string so we need to convert it into list
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

There is something we need to do before concatinating the columns,
we need to remove the extra spaces that come in names . For example,
Sam Worthington -> SamWorthington , Why is it needed?
The system will take both names Sam and Worthington as different entites while it is the name of a single entity,
same ways there could be many sam like Sam Mendes , so it will get confused which sam tag to choose , it could be
possible we need movies of sam worthington but got recommended of sam mendes . 
#In short the summary is to remove the space that comes in names

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title' , 'tags']].copy()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# now we need to do text vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
#stemming
#for example [loving , love , loved] => [love , love , love]

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

# cosine distance for cosine similarity


```python
# similarity is inverse of distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
# recommend top 5 movies whose similarity score is high
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)) , reverse = True , key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
